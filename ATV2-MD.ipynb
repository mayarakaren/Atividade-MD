A mineração de dados é uma etapa essencial para extrair informações valiosas de grandes volumes de dados. Neste caso, trabalhamos com um dataset que contém informações de vendas, incluindo produtos vendidos, quantidades, preços unitários e datas de venda. O objetivo foi realizar análises que destacassem os produtos mais vendidos e os que geraram maior receita. Para isso, utilizamos a biblioteca `pandas` para manipulação dos dados e `matplotlib` para visualização, permitindo uma interpretação clara e objetiva dos resultados.


## **1. Importações e Carregamento do Arquivo CSV**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar o arquivo CSV
file_path = 'vendas_5000_registros.csv'  
data = pd.read_csv(file_path)
data.head(20)
```

Nesta etapa, o arquivo CSV é carregado em um DataFrame utilizando a biblioteca `pandas`.

* O caminho do arquivo é definido na variável `file_path`. Você deve substituir pelo local correto do seu arquivo.
* O DataFrame resultante contém as informações das vendas, que serão usadas nas análises.


## **2. Agrupar por Nome do Produto e Somar as Quantidades Vendidas**

```python
# **1. Agrupar por nome do produto e somar as quantidades vendidas**
quantidade_por_produto = data.groupby('nome_produto')['quantidade'].sum().sort_values(ascending=False)
```

* O método `groupby` agrupa os dados pelo campo `nome_produto`.
* Em seguida, a função `sum` soma as quantidades vendidas de cada produto.
* Por fim, os resultados são ordenados em ordem decrescente com `sort_values(ascending=False)`, destacando os produtos mais vendidos.

## **3.  Gerar o Gráfico de Barras dos Produtos Mais Vendidos**

```python
# **2. Gerar o gráfico de barras dos produtos mais vendidos**
plt.figure(figsize=(10, 4))
quantidade_por_produto.head(10).plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Top 10 Produtos Mais Vendidos', fontsize=16)
plt.xlabel('Nome do Produto', fontsize=12)
plt.ylabel('Quantidade Vendida', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

* Um gráfico de barras é gerado utilizando os 10 produtos mais vendidos (`head(10)`).
* A biblioteca `matplotlib` é usada para criar e personalizar o gráfico:
    * `figsize`: Define o tamanho da figura.
    * `title`, `xlabel`, `ylabel`: Adicionam título e rótulos ao gráfico.
    * `xticks`: Rotaciona os nomes dos produtos para facilitar a leitura.
    * `grid`: Adiciona uma grade para facilitar a visualização dos valores.
* O gráfico mostra claramente quais produtos lideram em vendas.

## **4.  Calcular a Receita Gerada por Cada Produto**

```python
# **3. Calcular a receita gerada por cada produto**
data['receita'] = data['quantidade'] * data['preco_unitario']
receita_por_produto = data.groupby('nome_produto')['receita'].sum().sort_values(ascending=False)
```

* Uma nova coluna chamada `receita` é criada no DataFrame, calculando o total gerado por cada venda como `quantidade * preco_unitario`.
* Assim como na etapa anterior, os dados são agrupados por `nome_produto` e somados.
* O resultado é ordenado em ordem decrescente para destacar os produtos com maior receita.

## **5.  Gerar o Gráfico de Barras das Receitas por Produto**

```python
# **4. Gerar o gráfico de barras das receitas por produto**
plt.figure(figsize=(10, 4))
receita_por_produto.head(10).plot(kind='bar', color='orange', edgecolor='black')
plt.title('Top 10 Produtos com Maior Receita', fontsize=16)
plt.xlabel('Nome do Produto', fontsize=12)
plt.ylabel('Receita Total (R$)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

* O gráfico mostra os 10 produtos com maior receita total.
* Personalizações são semelhantes ao gráfico anterior, mas o foco aqui é na receita (coluna `receita`).

A análise revelou os produtos que se destacam tanto em termos de quantidade vendida quanto em geração de receita. A visualização por meio de gráficos de barras foi fundamental para identificar rapidamente quais itens são mais populares e lucrativos. Esses insights podem ser utilizados para tomar decisões estratégicas, como ajustar estoques, criar campanhas promocionais ou priorizar produtos de maior impacto financeiro. Com isso, a mineração de dados demonstra seu valor como uma ferramenta indispensável para a gestão e otimização de vendas.
